### Testy statystyczne

```python
from pingouin import multivariate_normality
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_prep = pd.read_csv("gasometry_prepared_data.csv")
cols=["BETET",	"CO2TET",	"HCO3ACTE",	"HCO3STTE",	"O2SATTET",	"O2TET", "IONH"]
my_blue="#0064B2"
my_red="#D61600"
```

```python
multivariate_normality(df_prep[cols], alpha=.05) 
```

```python
df_line=df_prep[[ "BADANIE_NR", "IONH","ZGON"]].copy()
df_line["ZGON"].replace({0: "zgon=nie", 1: "zgon=tak"}, inplace=True)
# sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
fig1=sns.lineplot(data=df_line, x="BADANIE_NR", y='IONH', hue="ZGON", ci='sd', palette={'zgon=nie': my_blue, 'zgon=tak': my_red}),

plt.xlabel('measurement number')
plt.ylabel('[H$^{\plus}$]')
plt.xlim(0,50)
plt.xticks(ticks=[1,10,20,30,40,50])
plt.legend(title="")
```
